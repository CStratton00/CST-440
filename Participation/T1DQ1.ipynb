The code snippets below depict the basic tools of networkx and how to create a bipartite graph. The dataset used has data collected from 246 physicians from Illinois about whether they trust other physicians in their profession. Each node is a physician and each edge the the amount of trust the first physician has in the second. 

```python
# import the networkx library
import networkx as nx
import pandas as pd

# create a graph
G = nx.Graph()

# load in the data set
df = pd.read_csv("bipartite_data.csv", sep=' ')
df.columns = ['doctor1', 'doctor2']

# check the data
df.head()
```

```python
# add in the edges
for idx in df.index:
    G.add_edge(df.loc[idx]['doctor1'], df.loc[idx]['doctor2'])

# draw the graph
nx.draw(G, with_labels=True)
```

```python

# create better visualization using NetworkX Visuals
from nxviz import CircosPlot

# draw will give an error but there is no actual error present, the code was referenced from their documentation
c = CircosPlot(G)
c.draw()
```
