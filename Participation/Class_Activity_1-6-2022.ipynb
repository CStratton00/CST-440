## Create a Scenario and Represent it with an Adjacency Matrix
In the example below, multiple cities from my home down and their distances between them were used to create an adjacency matrix and bipartite graph. 

```python
# import the networkx library
import networkx as nx

# create a list of cities
cities = ['Springfield', 'Decatur', 'Champaign', 'Indianapolis', 'Bloomington', 'Chicago', 'Rockford']

# create a graph and add the cities as nodes
g = nx.Graph()
g.add_nodes_from(cities)

# added edges to the cities in the graph
g.add_edge('Springfield', 'Decatur', weight=40)
g.add_edge('Springfield', 'Bloomington', weight=68)
g.add_edge('Decatur', 'Bloomington', weight=46)
g.add_edge('Decatur', 'Champaign', weight=49)
g.add_edge('Champaign', 'Bloomington', weight=49)
g.add_edge('Champaign', 'Chicago', weight=135)
g.add_edge('Champaign', 'Indianapolis', weight=120)
g.add_edge('Indianapolis', 'Chicago', weight=182)
g.add_edge('Bloomington', 'Rockford', weight=133)
g.add_edge('Bloomington', 'Chicago', weight=132)
g.add_edge('Chicago', 'Rockford', weight=89)

# draw the graph
nx.draw(g, with_labels=True)
```

```python
# create an adjacency matrix
mat = nx.adjacency_matrix(g)
mat.todense()
```
