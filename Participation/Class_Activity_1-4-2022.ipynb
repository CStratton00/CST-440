## Create a Bipartite Network and Graph
Bipartite networks are graphs whose vertices can be divided into two disjoint and independent sets. In these two sets, A and B, every edge connects a vertex in A to a vertex in B. In more general terms, each node represents a subject, and the edges between the nodes connect subjects together, or rather their interaction together. 

Some characteristics of a bipartite graph are as follows. Bipartite graphs can only be an even cycle, has an odd number of connections with minimal subsets of edges, and the spectrum of the graph is symmetric. 

See below my implementation of a bipartite network using python and networkx. 

### Basic Bipartite Graph
The code and graph below depict a basic bipartite network where there is an even number of nodes with a minimal number of connections. 

```python
# import the networkx library
import networkx as nx

# create a graph
G = nx.Graph()

# add nodes
G.add_nodes_from([1, 2, 3, 4], bipartite=0)
G.add_nodes_from(['a', 'b', 'c'], bipartite=1)

# add edges
G.add_edges_from([(1, "a"), (1, "b"), (2, "b"), (2, "c"), (3, "c"), (4, "a")])

# draw the graph
nx.draw(G, with_labels=True)
```

### Randomly Generated Bipartite Network
The code and graph below generate a random bipartite network. 

```python
# import the bipartite package
from networkx.algorithms import bipartite

# create a randomized bipartite graph
H = bipartite.random_graph(10, 5, 0.5)

# draw the graph
nx.draw(H, with_labels=True)
```
