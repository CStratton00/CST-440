# Introduction
In this project, we will analyze a network between companies to depict various types of inter-organization relationships. The objective is to detect potential for alliances, the impact of interlocking boards, and attempt to identify strategic opportunities and challenges.

# Code
We first imported the necessary libraries that are used for this project. We then printed the dataset as a table to show the relationship between companies. An important note about the data, one represents a financial connection between two companies and zero means that the two companies are not connected in any way or in some cases it is used to show a company being related to itself. The table was then plotted as a bipartite graph to be able to gain a visual understanding of the table for better analysis. 

```python
# import libraries
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Read in the data
df = pd.read_csv('FinancialTies.csv', index_col=0)
df.head()
```

```python
# create adjacency list
G = nx.from_pandas_adjacency(df)

# plot the adjacency list
plt.figure(figsize=(20,20))
nx.draw(G, with_labels=True)
```
