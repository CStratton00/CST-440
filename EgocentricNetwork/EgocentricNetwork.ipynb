```python
#Libraries used for this project
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("Github_following.csv")
df.head()
```

```python
followers_graph = nx.from_pandas_edgelist(df, source="V1", target="V2"
type(followers_graph)
```

```python
followers_graph.edges()
```

```python
nx.draw(followers_graph,with_labels = True)
```

```python
nx.degree(followers_graph)
```

```python
plt.figure(figsize=(25,25))
nx.draw_networkx(followers_graph)
plt.show()
```

```python
#degree of centrality
nx.degree_centrality(followers_graph)
```

```python
most_influential = nx.degree_centrality(followers_graph)
for w in sorted(most_influential, key = most_influential.get, reverse = True):
        print(w, most_influential[w])
```

```python
#Most Important Connection
most_important_link = nx.eigenvector_centrality(followers_graph)
for w in sorted(most_important_link, key=most_important_link.get, reverse = True):
    print(w, most_important_link[w])
```

```python
#Closeness
best_connector = nx.betweenness_centrality(followers_graph)
for w in sorted(best_connector, key=best_connector.get, reverse = True):
    print(w, best_connector[w])
```

```python
#Finding subgroups
groupCol = nx.bfs_tree(followers_graph, "CStratton00")
plt.figure(figsize=(20,20))
nx.draw_networkx(groupCol)
plt.show()
```

```python
#Finding subgroups
groupMXM = nx.bfs_tree(followers_graph, "mxmnci")
plt.figure(figsize=(20,20))
nx.draw_networkx(groupMXM)
plt.show()
```
